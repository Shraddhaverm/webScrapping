```python
# Cleaning using Stop Words Lists
nltk.download('stopwords')

# Load the positive and negative word lists
positive_words = set(open("positive_words.txt").read().splitlines())
negative_words = set(open("negative_words.txt").read().splitlines())

# Load the Excel file containing URLs
input_file = "input.xlsx"
df = pd.read_excel(input_file)

# Create a directory to store the extracted articles
output_directory = "extracted_articles"
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Function to extract data from a single link
def extract_data_from_link(url, url_id):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        # Extract article title
        title = soup.find('title').text
        # Extract article text
        article_text = ""
        article_elements = soup.find_all(['p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6'])
        for element in article_elements:
            article_text += element.text + "\n"
        # Save the extracted text to a text file
        filename = f"{output_directory}/{url_id}.txt"
        with open(filename, 'w', encoding='utf-8') as file:
            file.write(title + "\n\n")
            file.write(article_text)
        print(f"Extracted article from {url} and saved to {filename}")
        return filename  # Return the filename for sentiment analysis
    else:
        print(f"Failed to retrieve content from URL: {url}")
        return None

# Function to clean text by removing stop words
def clean_text(text):
    # Tokenize the text
    tokens = text.split()
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    cleaned_tokens = [token for token in tokens if token.lower() not in stop_words]
    # Join the cleaned tokens back into text
    cleaned_text = ' '.join(cleaned_tokens)
    return cleaned_text

# Function to extract derived variables
def extract_variables(text):
    # Initialize variables
    positive_word_count = 0
    negative_word_count = 0
    
    # Tokenize the text
    tokens = text.split()
    
    # Count positive and negative words
    for token in tokens:
        if token.lower() in positive_words:
            positive_word_count += 1
        elif token.lower() in negative_words:
            negative_word_count += 1
    
    # Calculate sentiment score using TextBlob
    blob = TextBlob(text)
    sentiment_score = blob.sentiment.polarity
    
    return positive_word_count, negative_word_count, sentiment_score

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    # Extract data from the link
    filename = extract_data_from_link(url, url_id)
    
    if filename:
        # Read the extracted text from the file
        with open(filename, 'r', encoding='utf-8') as file:
            article_text = file.read()
        
        # Clean the text
        cleaned_text = clean_text(article_text)
        
        # Extract derived variables
        positive_word_count, negative_word_count, sentiment_score = extract_variables(cleaned_text)
        
        print(f"Derived variables for {url}: Positive Word Count = {positive_word_count}, Negative Word Count = {negative_word_count}, Sentiment Score = {sentiment_score}")
        
```

```python
# Analysis of readability
# Function to calculate Flesch Reading Ease score
def calculate_reading_ease(text):
    words = text.split()
    num_words = len(words)
    num_sentences = text.count('.') + text.count('!') + text.count('?')
    num_syllables = 0
    for word in words:
        num_syllables += syllable_count(word)
    score = 206.835 - 1.015 * (num_words / num_sentences) - 84.6 * (num_syllables / num_words)
    return score

# Function to calculate Flesch-Kincaid Grade Level
def calculate_grade_level(text):
    words = text.split()
    num_words = len(words)
    num_sentences = text.count('.') + text.count('!') + text.count('?')
    num_syllables = 0
    for word in words:
        num_syllables += syllable_count(word)
    score = 0.39 * (num_words / num_sentences) + 11.8 * (num_syllables / num_words) - 15.59
    return score

# Function to count syllables in a word (simplified method)
def syllable_count(word):
    # Simple method to count syllables in a word
    count = 0
    vowels = 'aeiouy'
    word = word.lower()
    if word[0] in vowels:
        count += 1
    for index in range(1, len(word)):
        if word[index] in vowels and word[index - 1] not in vowels:
            count += 1
    if word.endswith('e'):
        count -= 1
    if count == 0:
        count += 1
    return count

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    # Extract data from the link
    filename = extract_data_from_link(url, url_id)
    
    if filename:
        # Read the extracted text from the file
        with open(filename, 'r', encoding='utf-8') as file:
            article_text = file.read()
        
        # Clean the text
        cleaned_text = clean_text(article_text)
        
        # Calculate readability scores
        flesch_reading_ease = calculate_reading_ease(cleaned_text)
        flesch_kincaid_grade_level = calculate_grade_level(cleaned_text)
        
        print(f"Readability analysis for {url}:")
        print(f"Flesch Reading Ease Score: {flesch_reading_ease}")
        print(f"Flesch-Kincaid Grade Level: {flesch_kincaid_grade_level}")
```

```python
#Average Number of Words Per Sentence
# Function to calculate average number of words per sentence
def calculate_avg_words_per_sentence(text):
    sentences = nltk.sent_tokenize(text)
    total_words = 0
    total_sentences = len(sentences)
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
        total_words += len(words)
    avg_words_per_sentence = total_words / total_sentences
    return avg_words_per_sentence

# Iterate through each row in the DataFrame
for index, row in input_df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    # Extract data from the link
    filename = extract_data_from_link(url, url_id)
    
    if filename:
        # Read the extracted text from the file
        with open(filename, 'r', encoding='utf-8') as file:
            article_text = file.read()
        
        # Clean the text
        cleaned_text = clean_text(article_text)
        
        # Calculate average number of words per sentence
        avg_words_per_sentence = calculate_avg_words_per_sentence(cleaned_text)
        
        print(f"Average number of words per sentence for {url}: {avg_words_per_sentence}")
        
```

```python
# Function to count syllables in a word
def syllable_count(word):
    vowels = 'aeiouy'
    count = 0
    prev_char_is_vowel = False
    for char in word:
        if char.lower() in vowels:
            if not prev_char_is_vowel:
                count += 1
            prev_char_is_vowel = True
        else:
            prev_char_is_vowel = False
    return count

# Function to count complex words
def count_complex_words(text):
    words = nltk.word_tokenize(text)
    complex_word_count = 0
    for word in words:
        syllable_count_word = syllable_count(word)
        # Define a threshold for what constitutes a complex word
        if syllable_count_word >= 3:
            complex_word_count += 1
    return complex_word_count

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    # Extract data from the link
    filename = extract_data_from_link(url, url_id)
    
    if filename:
        # Read the extracted text from the file
        with open(filename, 'r', encoding='utf-8') as file:
            article_text = file.read()
        
        # Clean the text
        cleaned_text = clean_text(article_text)
        
        # Count complex words
        complex_word_count = count_complex_words(cleaned_text)
        
        print(f"Complex word count for {url}: {complex_word_count}")
```

```python
# Function to count total words
def count_total_words(text):
    words = nltk.word_tokenize(text)
    total_words = len(words)
    return total_words

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    # Extract data from the link
    filename = extract_data_from_link(url, url_id)
    
    if filename:
        # Read the extracted text from the file
        with open(filename, 'r', encoding='utf-8') as file:
            article_text = file.read()
        
        # Clean the text
        cleaned_text = clean_text(article_text)
        
        # Count total words
        total_words = count_total_words(cleaned_text)
        
        print(f"Total word count for {url}: {total_words}")
```

```python
# Function to count syllables in a word
def syllable_count(word):
    vowels = 'aeiouy'
    count = 0
    prev_char_is_vowel = False
    for char in word:
        if char.lower() in vowels:
            if not prev_char_is_vowel:
                count += 1
            prev_char_is_vowel = True
        else:
            prev_char_is_vowel = False
    return count

# Function to count syllables per word
def count_syllables_per_word(text):
    words = nltk.word_tokenize(text)
    syllables_per_word = {}
    for word in words:
        syllable_count_word = syllable_count(word)
        syllables_per_word[word] = syllable_count_word
    return syllables_per_word

# Iterate through each row in the DataFrame
for index, row in input_df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    # Extract data from the link
    filename = extract_data_from_link(url, url_id)
    
    if filename:
        # Read the extracted text from the file
        with open(filename, 'r', encoding='utf-8') as file:
            article_text = file.read()
        
        # Clean the text
        cleaned_text = clean_text(article_text)
        
        # Count syllables per word
        syllables_per_word = count_syllables_per_word(cleaned_text)
        
        print(f"Syllable count per word for {url}: {syllables_per_word}")
```

```python
# Function to count personal pronouns
def count_personal_pronouns(text):
    personal_pronouns = ["i", "me", "my", "mine", "myself",
                         "you", "your", "yours", "yourself", "yourselves",
                         "he", "him", "his", "himself",
                         "she", "her", "hers", "herself",
                         "it", "its", "itself",
                         "we", "us", "our", "ours", "ourselves",
                         "they", "them", "their", "theirs", "themselves"]
    words = nltk.word_tokenize(text)
    personal_pronoun_count = 0
    for word in words:
        if word.lower() in personal_pronouns:
            personal_pronoun_count += 1
    return personal_pronoun_count

# Iterate through each row in the DataFrame
for index, row in input_df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    # Extract data from the link
    filename = extract_data_from_link(url, url_id)
    
    if filename:
        # Read the extracted text from the file
        with open(filename, 'r', encoding='utf-8') as file:
            article_text = file.read()
        
        # Clean the text
        cleaned_text = clean_text(article_text)
        
        # Count personal pronouns
        personal_pronoun_count = count_personal_pronouns(cleaned_text)
        
        print(f"Personal pronoun count for {url}: {personal_pronoun_count}")
```

```python
# Function to calculate average word length
def calculate_avg_word_length(text):
    words = nltk.word_tokenize(text)
    total_word_length = sum(len(word) for word in words)
    total_words = len(words)
    avg_word_length = total_word_length / total_words
    return avg_word_length

# Iterate through each row in the DataFrame
for index, row in df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    # Extract data from the link
    filename = extract_data_from_link(url, url_id)
    
    if filename:
        # Read the extracted text from the file
        with open(filename, 'r', encoding='utf-8') as file:
            article_text = file.read()
        
        # Clean the text
        cleaned_text = clean_text(article_text)
        
        # Calculate average word length
        avg_word_length = calculate_avg_word_length(cleaned_text)
        
        print(f"Average word length for {url}: {avg_word_length}")
```

```python
# Function to calculate percentage of complex words
def calculate_percentage_complex_words(text):
    total_words = len(nltk.word_tokenize(text))
    complex_word_count = count_complex_words(text)
    percentage_complex_words = (complex_word_count / total_words) * 100
    return percentage_complex_words

# Iterate through each row in the DataFrame
for index, row in input_df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    # Extract data from the link
    filename = extract_data_from_link(url, url_id)
    
    if filename:
        # Read the extracted text from the file
        with open(filename, 'r', encoding='utf-8') as file:
            article_text = file.read()
        
        # Clean the text
        cleaned_text = clean_text(article_text)
        
        # Calculate percentage of complex words
        percentage_complex_words = calculate_percentage_complex_words(cleaned_text)
        
        print(f"Percentage of complex words for {url}: {percentage_complex_words}%")
```
